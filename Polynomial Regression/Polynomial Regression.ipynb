```r
library(ggplot2)
plot1 <- ggplot(data=Data2, aes(x=Time, y=Counts))

#scatter plot of the data along with regression line
plot1 +geom_point(shape=16, color="red", alpha=0.6)+ 
      ggtitle("Count vs Time, (order1)")+
      geom_smooth(method=lm, se=F)
```

```r
plot3 <- plot2 + geom_smooth(method=lm, se=F)
plot3 
```

```r
## Linear Regression
## Building the model with Counts as dependent variable

model1 <- lm(formula=Counts ~ Time, data=Data2)
summary(model1)

# The model explains over 74% of the variance and has highly significant coefficients 
# for the intercept and the independent variable and also a highly significant overall
# model p-value
```

```r
predict1 <- predict(model1, Data2)

#install.packages("Metrics")
library(Metrics)
mse(Data1$Counts, predict1)
```

## (c) Polynomial Regression

```r
## building models with higher order polynomials
## predicting the same data with model built

Compare <- Data2
plot2
for(degree in 1:21) {
  # Building the model on complete data set with higher order polynomials
  fm <- lm(Counts ~ poly(Time, degree), data = Data2)
  assign(paste("model", degree, sep = "."), fm)

  # Predicting the data on the original dataset
  pr <- as.numeric(predict(fm, Data2))
  assign(paste("predict", degree, sep="."), pr)

  Compare <- cbind(Compare, pr)
  
  plot4 <- plot2 + geom_line(aes(Data2$Time,pr),color=degree)+ggtitle(paste("Time vs Counts, degree =",degree))
  print(plot4)
  
  print(paste("ORDER", degree, "- R Squared = ", summary(fm)$adj.r.squared,
              "& MSE =",(mse(Data2$Counts, pr) )))

  }


# The model with 2nd order polynomial is underfitting 
# while the the model with 13th order polynomial is overfitting
```

```r
for(degree in 1:17) {
  fm <- lm(Counts ~ poly(Time, degree), data = train)
  assign(paste("nmodel", degree, sep = "."), fm)
  
  pr <- as.numeric(predict(fm, pred))
  assign(paste("npredict", degree, sep="."), pr)
  
  Pred.Compare <- cbind(Pred.Compare, pr)
  
  print(paste("ORDER", degree, "- R Squared = ", summary(fm)$adj.r.squared,
              "& MSE =",(mse(train$Counts, pr) )))

}

# The model performing well on unseen data
```
